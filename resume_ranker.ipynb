```python
# Installing and import libaries
!pip install PyPDF2

import os
import PyPDF2
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#PDF text extract function
def extract_text_from_pdf(path):
    text = ""
    with open(path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() if page.extract_text() else ""
    return text
```

```python
#Load Resume
resume_folder = "/content"

resumes = []
names = []

for file in os.listdir(resume_folder):
    if file.endswith(".pdf"):
        text = extract_text_from_pdf(os.path.join(resume_folder, file))
        resumes.append(text)
        names.append(file)

print("Total resumes:", len(resumes))
```

```python
# Load job Description
# Make sure to upload 'job_description.txt' to the '/content/' directory.
# If you don't have one, you can create a dummy file for testing purposes:
with open('/content/job_description.txt', 'w') as f:
    f.write('This is a sample job description. We are looking for a candidate with strong Python skills and machine learning experience.')

with open("/content/job_description.txt", "r") as f:
    job_desc = f.read()

print("Job Description Loaded ✅")
print(job_desc[:150])
```

```python
# TF-IDF vectorization
documents = resumes + [job_desc]

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
#Cosine Similarity
scores = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])
```

```python
# Ranking Output
ranking = np.argsort(scores[0])[::-1]

print(" Resume Ranking:\n")

for i in ranking:
    print(f"{names[i]} --> Score: {scores[0][i]:.4f}")
```
